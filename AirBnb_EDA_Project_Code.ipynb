#### Chart - 1 : Distribution of Listing Prices (Box Plot)

````python
# Chart 1(a) - Distribution of Listing Prices (Boxplot)(Outlier Detection)

plt.figure(figsize=(10,6))
sns.boxplot(x=raw_df['price'])
plt.title('Distribution of Listing Prices')
plt.xlabel('Price ($)')
plt.show()
````

##### 1. Why did you pick the specific chart?

**Box plots are best for removing outliers, as outliers are typically defined as data points that fall outside the "whiskers" of the box plot.**

**The whiskers extend 1.5 times the interquartile range (IQR) from the quartiles (Q1 and Q3). This makes it easy to visually identify potential outliers.**

##### 2. What is/are the insight(s) found from the chart?

**The box plot is too skewed and this shows the presence of outliers in the price column.**

##### 3. Will the gained insights help creating a positive business impact?


**Yes, I will now be clearing the outliers**

````python
# Defining a function to remove the outliers from the 'price' variable using IQR method

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Removing the outliers and creating a new dataset named 'filtered_df'
filtered_df = remove_outliers(raw_df, 'price')

#Printing the new dataset
filtered_df
````

**The outliers have finally been removed ðŸ‘† and now the box plot will look reasonably more informative and correct to visualize.**

##### 4. Are there any insights that lead to negative growth? Justify with specific reason.

````python
# Chart 1(b) - Distribution of Listing Prices (Boxplot)(Insight Generation)

sns.set_style('whitegrid')

# Plotting the box-plot again after removal of outliers from the 'price' variable.

plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(x=filtered_df['price'], orient='h', color='skyblue')

# Getting the quartile values (For Annotation Purpose)
Q1 = filtered_df['price'].quantile(0.25) # Q1 is a point below which 25% of the 'price' data lies.
Q2 = filtered_df['price'].quantile(0.50) # Q2 (Median) is a point below which 50% of the 'price' data lies.
Q3 = filtered_df['price'].quantile(0.75) # Q3 is a point below which 75% of the 'price' data lies.

# Getting IQR and whisker bounds (For Annotation Purpose)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Getting the exact value of the points, where the whiskers are present (For Annotation Purpose)
min_value = filtered_df['price'][filtered_df['price'] >= lower_bound].min()
max_value = filtered_df['price'][filtered_df['price'] <= upper_bound].max()

# Annotating the extreme values of the whiskers
plt.text(min_value, -0.2, f'{min_value:.2f}', ha='center', va='top', color='red')
plt.text(max_value, -0.2, f'{max_value:.2f}', ha='center', va='top', color='red')

# Annotating the quartiles values
plt.text(Q1, -0.1, f'Q1: {Q1:.2f}', ha='center', va='top', color='blue') # For Q1
plt.text(Q2, -0.2, f'Q2: {Q2:.2f}', ha='center', va='top', color='red') # For Q2
plt.text(Q3, -0.3, f'Q3: {Q3:.2f}', ha='center', va='top', color='blue') # For Q3

plt.title('Distribution of Listing Prices')
plt.xlabel('Price ($)')
plt.show()
````

#### Chart - 2 : Price distribution over Room type (Box Plot)

````python
# Visualization Code

plt.figure(figsize=(10, 6))
color_palette1 = ['#540b0e', '#ff7d00', '#f94144']
box_plot = sns.boxplot(x='room_type', y='price', data=filtered_df, palette = color_palette1)
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')

# Getting the quartiles for each room type
room_types = filtered_df['room_type'].unique()
for room_type in room_types:
    data = filtered_df[filtered_df['room_type'] == room_type]['price']
    Q1 = data.quantile(0.25)
    Q2 = data.quantile(0.50)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Getting the whisker values
    min_value = data[data >= lower_bound].min()
    max_value = data[data <= upper_bound].max()

    # Annotating the quartiles
    plt.text(room_types.tolist().index(room_type), Q1, f'Q1: {Q1:.2f}', ha='center', va='top', color='blue')
    plt.text(room_types.tolist().index(room_type), Q2, f'Mean: {Q2:.2f}', ha='center', va='center', color='black')
    plt.text(room_types.tolist().index(room_type), Q3, f'Q3: {Q3:.2f}', ha='center', va='bottom', color='blue')

    # Annotating the whiskers
    plt.text(room_types.tolist().index(room_type), min_value, f'{min_value:.2f}', ha='right', va='top', color='green')
    plt.text(room_types.tolist().index(room_type), max_value, f'{max_value:.2f}', ha='left', va='top', color='green')

plt.show()
````
